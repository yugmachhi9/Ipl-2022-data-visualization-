### key player performance

1] most "player of the match " awards

```python
count = df['player_of_the_match'].value_counts().head(10)
count
```

```python
sns.barplot(x=count.values,y= count.index)
plt.title("top 10 player")
```

2] top scorrer

```python
df.head()
```

```python
high = df.groupby('top_scorer')['highscore'].sum().sort_values(ascending = False).head(3)
high
```

```python
sns.barplot(high)
```

2] Best bowling

```python

bowl = df.groupby('best_bowling')['highest_wicket'].sum().sort_values(ascending = False).head(10)
bowl.plot(kind = 'barh') # barh means barplot horizontal
```
